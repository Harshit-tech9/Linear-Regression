```python
X = df.drop('sales', axis=1)
```

```python
X.head()
```

```python
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
help(train_test_split)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42) #test_size ask for data to be left out and put it in test variable
```

```python
#"""random_state shuffle data so all the datas are perfectly shuffled and we get unbiased data in both train and test"""
```

```python
len(df)
```

```python
len(X_train)
```

```python
len(y_train)
```

```python
len(X_test)
```

```python
len(y_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
help(LinearRegression)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_test.head()
```

```python
X_test.head()
```

```python
test_predictions = model.predict(X_test) #y_test
```

```python
test_predictions
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
round(df['sales'].mean(), 2)
```

```python
sns.histplot(x = df['sales'], bins = 20)
plt.show()
```

```python
mean_absolute_error(y_test, test_predictions)
```

```python
# standard deviation
np.sqrt(mean_squared_error(y_test, test_predictions))
```

```python
test_residuals = y_test - test_predictions
```

```python
test_residuals
```

```python
sns.scatterplot(x=y_test, y = test_residuals, color = 'green')
plt.axhline(y=0, color='r', ls = '--')
plt.show()
```

```python
sns.displot(test_residuals, bins = 25, kde= True)
plt.show()
```

```python
import scipy as sp
```

```python
fig, ax = plt.subplots(figsize=(6,8), dpi = 100)
Q = sp.stats.probplot(test_residuals, plot = ax)
plt.show()
```

## Model Deployment

```python
final_model = LinearRegression()
```

```python
final_model.fit(X,y)
```

```python
final_model.coef_
```

```python
X.head()
```

```python
y_hat = final_model.predict(X)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3, figsize=(16,6))

axes[0].plot(df['TV'],df['sales'],'*', color = 'green')
axes[0].plot(df['TV'],y_hat,'o', color = 'red')
axes[0].set_ylabel('Sales')
axes[0].set_title('TV spend')

axes[1].plot(df['radio'],df['sales'],'o', color = 'orange')
axes[1].plot(df['radio'],y_hat,'o', color = 'red')
axes[1].set_ylabel('Radio Sales')
axes[1].set_title('Radio spend')

axes[2].plot(df['newspaper'],df['sales'],'o', color = 'purple')
axes[2].plot(df['newspaper'],y_hat,'o', color = 'red')
axes[2].set_ylabel('Newspaper Sales')
axes[2].set_title('Newspaper spend')
plt.tight_layout()

plt.show()
```
